### Loops
In computer programming, a loop is a sequence of instruction's that is continually repeated until
a certain condition is reached. Typically, a certain process is done, such as getting an item of
data and changing it.
Loops help us to perform code in data when a task has to be repeated several times. With the
use of loops, we can cut short those hundred lines of code to a few.

###  Real life Applications

1. traffic controller light
2. You put your favorite song on a repeat mode or play 5 times.
3. Software program in a mobile device allows user to unlock the mobile with 5 password attempts. After that it resets mobile device.
4. display menu/catlog for products on e-commerce

## 1- For Loop

A for loop in Python is a control flow statement that is used to repeatedly execute a group of
statements as long as the condition is satisfied.

Syntax :- For in __:

For loop takes first item of iterable object and runs operation/code on it, then to the second
item and so on until the condition is satisfied.

### 2- While loop( conditional controlled)


While loop continually executes the statements(code) as long as the given condition is TRUE. 
It first checks the condition and then jumps into the instructions.


syntax :- while __ :



while loop takes first item and then it checks whether condition is true or false, if its true than it
run the operation/code and proceeds to next item and follows this part untile condition become
false. if condition become false it came out of loop and stops the loop.


### Examples:-

```python
## print table based on specified range.
table num1,num2=input("enter the number").split(" ")
num1=int(num1)
num2=int(num2)
while(num1<=num2):
    print(num1)
    i=1
    while(i<=10):
        print(num1,"*",i,"=",num1*i)
        i=i+1
    num1=num1+1
    
    

    
```

```python
### First we take iterable
a=1 
while a<=5: # while loop runs until value of a is less or equal to 5
    print('hello')#it will print hello until condition is true and we got 5 hello
    a+=1 # a+=1 for increasing value of a
```

```python
# applying if else condition in while loop
a = 1
even_numbers = []
while a<=10: # loop runs until value of a is less or equal to 10
    if a%2==0: # while condition is true it runs if condition
        even_numbers.append(a)
    a+=1
print(even_numbers)
```

```python
# When an else statement is used along with a while loop,
# the control goes to the else statement when the condition is False.
a = 1
while a <= 4:
    print('Condition is TRUE')
    a+=1
else:
    print('Condition is FALSE')
```

### Loop Control Statements
Loop control statements are used to change the flow of execution. These can be used if you
wish to skip an iteration or stop the execution.
The three types of loop control statements are:

1) break statement

2) continue statement

3) pass statement

### Break
we have used break operation above in while loop. it's used to break operation in middle or
desied point

```python
# in for loop it have to print numbers form 1-10 as specified
for i in range(1, 11):
    if i == 5:
        #print("hello")
        break
    print(i)
# but we applied break function at five so it only gives number form 1-4
#else:
    #print(i)
```

### Continue
Continue statement is used to skip the current item when the condition is met and allows the
loop to continue with the next item.

```python
a=0
while a<10:
    a+=1
    if a==9:
        continue
# while loop is true for a less than or eqaul 10 but it skips 9
# because of continue operation
    print(a)
```

### Pass
Pass statement is used when we want to do nothing when the condition is met. It doesnt skip or
stop the execution, it just passes to the next item.

```python
for i in range(1, 11):
    if i == 7:
        pass
# as you can see it didn't stop or skip 7 in loop it just let it pass
    print(i)
```

### wap to check whether the given number id armstrong or not


Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

```python
num=int(input("enter the number"))
sum=0
temp=num
while(num!=0):
    r=num%10
    num=num//10
    sum=sum+r*r*r
print(sum)
if temp==sum:
    print("armstrong number")
else :
    print("not armstrong")
```
