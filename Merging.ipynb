### Merging is used when we need to join two dataframes having different columns and there is a column which is common in dataframes

### It combines the dataframe columns wise only

### while concat combines the dataframes column wise and rows wise. 






There are four basic ways to handle the merge 



1-inner (default)


2-left


3-right

4-outer



####  MERGE METHOD	JOIN NAME	DESCRIPTION
left	 	Use keys from left frame only


right		Use keys from right frame only


outer		Use union of keys from both frames


inner	     Use intersection of keys from both frames

### Merging a dataframe with one unique key combination


```python

import pandas as pd 
 
# Define a dictionary containing employee data 
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 
   
# Define a dictionary containing employee data 
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data1)
 
# Convert the dictionary into DataFrame  
df2 = pd.DataFrame(data2) 
  


```

```python
df1
```

```python
df2
```

```python
merge_left = pd.merge(df1, df2,how='inner',on='key')
merge_left
```

```python
merge_left = pd.merge(df1, df2,how='left',on='key')
merge_left
```

```python
merge_right = pd.merge(df1, df2,how='right',on='key')
merge_right
```

```python
merge_outer = pd.merge(df1, df2,how='outer',on='key')
merge_outer
```

### from above results: merging result is same in all inner,left,right ,outer  joins the key is common and same value in both dataframes

### inner join  : default join


Its main task is to combine the two DataFrames based on a join key and returns a new DataFrame. 


#### The returned DataFrame consists of only selected rows that have matching values in both of the original DataFrame.

```python
## merging the data frame 
merge_inner=pd.merge(df1,df2,how='inner',on='key')
merge_inner
```

```python
## merging the data frame to the left
merge_left=pd.merge(df1,df2,how='left',on='key') ## keys of left will come 
merge_left
```

```python
#### merging the data frame to the left
merge_left1=pd.merge(df2,df1,how='left',on='key')
merge_left1
```

```python
##### merging the data frame to the right
merge_right=pd.merge(df1,df2,how='right',on='key') ## keys of right df will come
merge_right
```

```python
##### merging the data frame to the right
merge_right1=pd.merge(df2,df1,how='right',on='key')
merge_right1
```

```python
merge_outer=pd.merge(df1,df2,how='outer',on='key') # #keys of both data frame will come
merge_outer
```

### merging on more than one keys: key1 and key2 both as below 

```python
df1 = pd.DataFrame({'key1': ['K0', 'K8', 'K2', 'K3'],
                    'key2': ['K10', 'K18', 'K12', 'K13'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                    'key2': ['K10', 'K11', 'K12', 'K13'],
                          'A': ['C0', 'C1', 'C2', 'C3'],
                          'B': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
df2
```

```python
merge=pd.merge(df1,df2,how="inner",on=["key1","key2"]) ## considering key1 and key2 both as same
merge
```

```python
merge=pd.merge(df1,df2,how="left",on=["key1","key2"])
merge
```

```python
merge=pd.merge(df1,df2,how="right",on=["key1","key2"])
merge
```

```python
merge=pd.merge(df1,df2,how="outer",on=["key1","key2"])
merge
```
